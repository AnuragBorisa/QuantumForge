```python
from qiskit_finance.circuit.library import LogNormalDistribution
import matplotlib.pyplot as plt
dist = LogNormalDistribution(num_qubits=3, mu=0.9, sigma=0.9, bounds=(0, 1))
x, y = dist.values, dist.probabilities  # PDF values
plt.plot(x, y)
plt.title("Log-Normal PDF (μ=0.5, σ=0.25)")
plt.xlabel("Value")
plt.ylabel("Probability Density")
plt.show()
```
