```python
import numpy as np

samples = np.load("data/raw_log_normal.npy")

```

```python
import matplotlib.pyplot as plt

num_bins = 50

plt.figure(figsize=(6,4))
plt.hist(samples,bins=num_bins,color="C0",edgecolor = "w")
plt.title("Linear-scale histogram of synthetic prices")
plt.xlabel("Price")
plt.ylabel("Count")
plt.savefig("figures/linear_hist.png")

plt.figure(figsize=(6,4))
plt.hist(samples, bins=num_bins, color="C0", edgecolor="w")
plt.yscale("log")             
plt.title("Log-y histogram (shows heavy tail)")
plt.xlabel("Price")
plt.ylabel("Log(Count)")
plt.savefig("figures/logy_hist.png")

```

```python
from scipy.stats import lognorm

plt.figure(figsize=(6,4))
count,bins_edge, _ = plt.hist(samples,
                              bins=num_bins,density = True, 
                              color="C0", edgecolor="w",
                              alpha=0.6)

x = np.linspace(bins_edge[0],bins_edge[-1],200)

from math import exp

mu    = np.log(80)
sigma = 0.5

pdf_values = lognorm.pdf(x,s=sigma,scale=exp(mu))

plt.plot(x, pdf_values, 'r-', lw=2, label="theoretical PDF")
plt.title("Empirical vs. Theoretical Log-Normal PDF")
plt.xlabel("Price")
plt.ylabel("Density")
plt.legend()
plt.savefig("figures/pdf_overlay.png")

```
