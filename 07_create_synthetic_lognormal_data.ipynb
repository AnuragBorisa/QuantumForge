```python
import numpy as np 

mu,sigma = np.log(80) , 0.5

samples = np.random.lognormal(mean=mu,sigma=sigma,size=50_000)
```

```python
import matplotlib.pyplot as plt 

plt.hist(samples,bins=100,density=True,alpha=0.6)
plt.xlabel('Value')
plt.ylabel('Density')
plt.title('Histogram of log-normal samples')
plt.show()
```

```python
from scipy.stats import lognorm

x = np.linspace(samples.min(), samples.max(), 500)
pdf = lognorm.pdf(x, s=sigma, scale=np.exp(mu))
plt.plot(x, pdf, 'r-', lw=2, label='theoretical PDF')
plt.hist(samples, bins=100, density=True, alpha=0.4)
plt.legend()
plt.show()
```

```python
plt.hist(samples, bins=100, density=True)
plt.xscale('log')
plt.yscale('log')
plt.xlabel('Value (log scale)')
plt.ylabel('Density (log scale)')
plt.show()
```

```python
sorted_s = np.sort(samples)
ecdf = np.arange(1, len(sorted_s)+1) / len(sorted_s)
plt.step(sorted_s, ecdf)
plt.xscale('log')
plt.xlabel('Value')
plt.ylabel('Empirical CDF')
plt.title('Empirical CDF (log scale x-axis)')
plt.show()
```
