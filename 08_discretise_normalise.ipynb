```python
import numpy as np
import matplotlib.pyplot as plt 


samples = np.load('data/raw_log_normal.npy')
print(samples)

num_bins = 2**3

counts ,bin_edges = np.histogram(samples,bins=num_bins)

print(counts,bin_edges)

p = counts/counts.sum()

np.save('data/log_normal_bins.npy', bin_edges)
np.save('data/log_normal_p.npy',     p)

plt.bar(range(num_bins), p, width=1.0)
plt.xlabel('Bin index (0 … 7)')
plt.ylabel('Probability')
plt.title('Discretised Log-Normal → 8-entry p')
plt.tight_layout()
plt.show()
```
